### Conclusion 2.a

```python
#stacked barplot of diffrences between Older and Younger.
categories = ["please others", "accepting pos feedback or criticism", "comfort zone", "confident"]

custom_colors = ['red', '#F3C4FB', '#FFDFBA', '#D3FFCE', 'green']
titles = [
    "i dont try to please other",
    "do you have trouble accepting positive feedback or criticism?",
    "you love to try new things or come out of your comfort zone easily",
    "you consider yourself confident"
]
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
axes = axes.flatten()

for i, category in enumerate(categories):
    
    older_counts = df[df['age'] == 'Older'][category].value_counts(normalize=True).sort_index() * 100
    younger_counts = df[df['age'] == 'Younger'][category].value_counts(normalize=True).sort_index() * 100
    combined_df = pd.DataFrame({
        'Older': older_counts,
        'Younger': younger_counts
    }).fillna(0)

    combined_df = combined_df.transpose()
    combined_df.plot(kind='barh', stacked=True, ax=axes[i], color=custom_colors)

    for container in axes[i].containers:
        axes[i].bar_label(container, fmt='%.1f%%', label_type='center')

    axes[i].set_title(titles[i], fontsize=17)
    axes[i].set_yticklabels(['Older', 'Younger'], fontsize=14)
    axes[i].legend(title='Vote Category', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

4 components (questions) in which there are large differences (a double-digit number in percentage) between the older group and the younger ones that affect the components of the self-esteem that each group builds.

### Conclusion 3,a

```python
# an averge of each person generel confidence.
confidence_columns = [
    "comfort zone", "confident", "feel control", "compare", 
    "doubt yourself", "accepting pos feedback or criticism", 
    "fear of failure", "please others"]

df['average_confidence'] = df[confidence_columns].mean(axis=1)
analysis_df = df.sort_values(by='average_confidence').reset_index(drop=True)
```

```python
# Creating the line graph
plt.figure(figsize=(10, 6))
plt.plot(analysis_df.index, analysis_df['average_confidence'], marker='o', linestyle='-', color='b')
plt.title('Average Confidence of society')
plt.xlabel('subject')
plt.ylabel('Average Confidence Score')
plt.xticks(range(len(analysis_df)))
plt.show()
```

We created a type of self-confident person by combining some questions asked about the person who answered the survey. We created a graph that represents the distribution of all the people who answered the survey. The sample shows the average of each person who answered the survey.
We saw that our sample is distributed very homogeneously when there is a little at each end and this shows a very logical distribution of human society.

```python
# Classifying respondents based on average confidence scores.
def classify_confidence(average_confidence):
    if average_confidence >= 3.5:
        return 'High_confidence'
    elif average_confidence >= 2.75:
        return 'Medium_confidence'
    else:
        return 'Low_confidence'
    
df['confidence_level'] = df['average_confidence'].apply(classify_confidence)
```

```python
# The distribution of all people in relation to self-confidence.
fig, ax = plt.subplots()

df['confidence_level'] = pd.Categorical(df['confidence_level'], categories=['Low_confidence', 'Medium_confidence', 'High_confidence'], ordered=True)
sns.histplot(x='confidence_level', data=df, hue='confidence_level', palette={'Low_confidence': '#ff9999', 'Medium_confidence': '#ffff99', 'High_confidence': '#99ff99'})

fig.subplots_adjust(left=1.5, right=3, top=1, bottom=0.2)
plt.show()
```

We created three groups with a high self-confidence group, a medium self-confidence group and a low self-confidence group.
The vast majority are somewhere in between.

```python
#Decrease to resolutions of specific groups. Here we tested the boys versus the girls.
df['confidence_level'] = pd.Categorical(df['confidence_level'], categories=['Low_confidence', 'Medium_confidence', 'High_confidence'], ordered=True)

df['confidence_code'] = df['confidence_level'].cat.codes

palette = {
    'Female': 'pink',
    'Male': 'lightblue',
    'Low_confidence': '#ff9999',  
    'Medium_confidence': '#ffff99',  
    'High_confidence': '#99ff99'  
}

plt.figure(figsize=(10, 6))
sns.histplot(x='confidence_level', data=df, hue='gender', multiple='dodge', palette=palette, shrink=0.8, element='bars')

plt.show()
```

We examined how it is divided into groups that we have already examined above, and this time we are not dealing with specific questions, but rather with a type of a person we created that includes several qualities together that create confidence. We look at it from a macro perspective.
We see that even though there are fewer boys than girls, there are still more boys than girls who define themselves as highly self-confident.

```python
#To see the distribution of boys relative to girls in the graph is more understandable and to see that the average of girls is lower than the average of boys.
df['confidence_level'] = pd.Categorical(df['confidence_level'], categories=['Low_confidence', 'Medium_confidence', 'High_confidence'], ordered=True)

df['confidence_code'] = df['confidence_level'].cat.codes

palette = {
    'Female': 'pink',
    'Male': 'lightblue'}
plt.figure(figsize=(10, 6))

for gender, color in zip(['Female', 'Male'], ['pink', 'lightblue']):
    subset = df[df['gender'] == gender]
    sns.kdeplot(data=subset, x='confidence_code', color=color, label=f'{gender} Distribution', linewidth=2, fill=True)
    
    # Calculate and plot the mean
    mean_value = subset['confidence_code'].mean()
    plt.axvline(mean_value, color=color, linestyle='--', linewidth=2)
    plt.text(mean_value, plt.ylim()[1] * 0.8, f'{gender} Mean', color=color, ha='center')

plt.title('Confidence Level Distribution by Gender')
plt.xlabel('Confidence Level')
plt.ylabel('Density')
plt.legend(title='Gender')
plt.grid(True)

plt.xticks(ticks=[0, 1, 2], labels=['Low_confidence', 'Medium_confidence', 'High_confidence'])

plt.show()
```

We showed in another way the distribution between boys and girls, when here you can see very nicely the dotted line that shows the average self-confidence of each gender according to the type of person we created.
